```python
 from nltk import pos_tag
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
morph = []
tmp_list = []
tmp_tagged_list = []
for lyric in music_lyrics_string:
  tmp = word_tokenize(lyric)
  tmp_tagged = pos_tag(tmp)
  tmp_tagged_list.append(tmp_tagged)
```

(words in dataset are already in lowercase)

```python
print(tmp_tagged_list[:1])
```

```python
with open("tmp_tagged_list.txt", "w") as output:
    output.write(str(tmp_tagged_list))
```

```python
nltk.download('stopwords')
stopwords_set = stopwords.words('english')
stopwords_set.extend(["would", "could", "one", "may", "im", "ive", "dont", "youre", "oh", "cause", "yeah", "yeah yeah", "oh oh", "da da", "wan na", "im gon"])
```

```python
print(stopwords_set)
```

```python
morph = []
for i in tmp_tagged_list:
    for j in i:
        if (j[1] in ['JJ', 'JJR', 'JJS', 'NN', 'NNP', 'NNS', 'RBR', 'RBS']):
            if j[0] not in stopwords_set:
#                 print(j[0])
                morph.append(j[0])
                                                
```

```python
morph_joined = ' '.join(morph)
```

```python
wordcloud = WordCloud(max_words=10, font_path='./MaruBuri-Regular.otf', max_font_size=100, background_color='white').generate(morph_joined)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(wordcloud, interpolation='lanczos')
plt.axis('off')
plt.show()
```

```python
musicAllAround = pd.read_csv("lyrics-data.csv")
```

```python
musicAllAroundLyrics = musicAllAround["Lyric"]
```

```python
print(musicAllAroundLyrics.head())
```

```python
with open("musicAllAroundLyrics.txt", "w") as output:
#     output.write(str(music["Lyrics"]))
#     output.write(str(music_lyrics_string))
    for i in range(len(musicAllAround["Lyric"])):
        output.write(str(musicAllAround["Lyric"][i]) + "\n")
```

```python
def is_keyword_in_lyric(lyric):
    count = 0
    for word in keywords:
        if word in lyric:
            count = count +1
    if count >= 1:
        return True
    else:
        return False
```

```python
def is_keyword_in_lyric3(lyric):
    count = 0
    for word in keywords:
        if word in lyric:
            count = count +1
    if count >= 3:
        return True
    else:
        return False
```

```python
t = open("musicAllAroundLyrics.txt", "r")
f = open("musicAllAroundLyrics_modified.txt", "w")
for line in t.readlines():
    line = line.strip()
    if is_keyword_in_lyric3(line):
#         t.append('\n',1)
        f.writelines("__label__1\t" + line)
    else: 
        f.writelines("__label__0\t" + line)
f.close
        
    
```

```python
t = open("music_lyrics_string.txt", "r")
f = open("music_lyrics_string_modified.txt", "w")
for line in t.readlines():
    line = line.strip()
    if is_keyword_in_lyric3(line):
#         t.append('\n',1)
        f.writelines("__label__1\t" + line)
    else: 
        f.writelines("__label__0\t" + line)
f.close
        
```
